```python
import pandas as pd

# Load the dataset
file_path = "/content/original 1000 samples.xlsx"
df = pd.read_excel(file_path)

# Inspect required columns
required_cols = ["organism group", "assembly"]
missing = [c for c in required_cols if c not in df.columns]
if missing:
    raise ValueError(f"Missing required columns: {missing}")

# Drop rows without assembly accessions
df = df.dropna(subset=["assembly"])

# Get first 1000 unique organisms
unique_organisms = df["organism group"].drop_duplicates().iloc[:1000]

# Filter dataframe to those organisms
filtered_df = df[df["organism group"].isin(unique_organisms)]

# Select relevant columns
assemblies_df = filtered_df[
    ["organism group", "assembly"]
].drop_duplicates()

# Reset index
assemblies_df = assemblies_df.reset_index(drop=True)

# Save output for downstream processing
output_path = "/content/organisms_1_to_1000_assemblies.csv"
assemblies_df.to_csv(output_path, index=False)

print(f"‚úÖ Selected assemblies saved to: {output_path}")
print(f"üìä Total organisms: {assemblies_df['organism group'].nunique()}")
print(f"üß¨ Total assemblies: {assemblies_df.shape[0]}")
```

```python
import os
import pandas as pd
import subprocess
from tqdm import tqdm

# Load assembly list
df = pd.read_csv("/content/all_1000_assemblies.csv")

assemblies = (
    df["assembly"]
    .dropna()
    .astype(str)
    .str.strip()
    .unique()
    .tolist()
)

print(f"üß¨ Total assemblies to download: {len(assemblies)}")

output_dir = "/content/genomes"
os.makedirs(output_dir, exist_ok=True)

failed = []

for acc in tqdm(assemblies):
    try:
        zip_path = f"{output_dir}/{acc}.zip"

        subprocess.run(
            [
                "datasets", "download", "genome", "accession", acc,
                "--filename", zip_path,
                "--include", "genome"
            ],
            check=True,
            stdout=subprocess.DEVNULL,
            stderr=subprocess.DEVNULL
        )

        subprocess.run(
            ["unzip", "-o", zip_path, "-d", f"{output_dir}/{acc}"],
            check=True,
            stdout=subprocess.DEVNULL,
            stderr=subprocess.DEVNULL
        )

        os.remove(zip_path)

    except Exception:
        failed.append(acc)

if failed:
    with open("/content/failed_assemblies.txt", "w") as f:
        f.write("\n".join(failed))

print("‚úÖ Genome download completed")
print(f"‚ùå Failed downloads: {len(failed)}")
```

```python
import pandas as pd
import re

# Load raw dataset
df = pd.read_excel("/content/original 1000 samples.xlsx")

# Clean column names
df.columns = df.columns.str.strip()

# Column names (exact from your file)
isolate_col = "isolaote identifier"
organism_col = "organism group"
assembly_col = "assembly"
source_col = "isolation source"
type_col = "isolation type of pathogen"
ast_col = "AST phenotype"

# Keep only rows with assemblies
df = df[df[assembly_col].notna()].copy()

print("Rows with assemblies:", df.shape[0])

# Parse antibiotic and phenotype
def parse_ast(value):
    if pd.isna(value):
        return None, None

    value = str(value)

    phenotype = None
    for p in ["Resistant", "Intermediate", "Susceptible"]:
        if p.lower() in value.lower():
            phenotype = p
            break

    antibiotic = re.sub(
        "(?i)resistant|intermediate|susceptible|[-:]", "", value
    ).strip()

    if not antibiotic or not phenotype:
        return None, None

    return antibiotic, phenotype

parsed = df[ast_col].apply(parse_ast)
df["antibiotic"] = parsed.apply(lambda x: x[0])
df["phenotype"] = parsed.apply(lambda x: x[1])

# Drop rows without valid AST
df = df.dropna(subset=["antibiotic", "phenotype"])

# Aggregate per isolate (GENOME-LEVEL)
rows = []

for isolate_id, g in df.groupby(isolate_col):
    rows.append({
        "isolate_id": isolate_id,
        "organism": g[organism_col].iloc[0],
        "assembly": g[assembly_col].iloc[0],
        "isolation_source": g[source_col].iloc[0],
        "isolation_type": g[type_col].iloc[0],
        "resistant_antibiotics": "; ".join(
            sorted(g[g["phenotype"] == "Resistant"]["antibiotic"].unique())
        ),
        "intermediate_antibiotics": "; ".join(
            sorted(g[g["phenotype"] == "Intermediate"]["antibiotic"].unique())
        ),
        "susceptible_antibiotics": "; ".join(
            sorted(g[g["phenotype"] == "Susceptible"]["antibiotic"].unique())
        ),
    })

clean_df = pd.DataFrame(rows)

# Save clean file
output_path = "/content/clean_771_genome_level_AST.csv"
clean_df.to_csv(output_path, index=False)

print("‚úÖ Clean genome-level AST file created")
print("üìÅ Saved to:", output_path)
print("üß¨ Total genomes:", clean_df.shape[0])
print("üß¨ Unique assemblies:", clean_df['assembly'].nunique())
```

```python
from Bio import SeqIO
import os

# QC parameters
MIN_CONTIG_LEN = 500
MAX_N_PERCENT = 5.0

# Input and output directories
INPUT_DIR = "/content/genomes"
OUTPUT_DIR = "/content/genomes_qc"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Logging
qc_log = []

def n_percentage(seq):
    return (seq.upper().count("N") / len(seq)) * 100

# Process each assembly folder
for root, _, files in os.walk(INPUT_DIR):
    for file in files:
        if file.endswith(".fna"):
            input_fasta = os.path.join(root, file)
            assembly_id = file.split(".")[0]
            output_fasta = os.path.join(OUTPUT_DIR, f"{assembly_id}.fna")

            kept_contigs = []
            removed_contigs = 0

            for record in SeqIO.parse(input_fasta, "fasta"):
                seq_len = len(record.seq)
                n_pct = n_percentage(str(record.seq))

                if seq_len >= MIN_CONTIG_LEN and n_pct <= MAX_N_PERCENT:
                    kept_contigs.append(record)
                else:
                    removed_contigs += 1

            if kept_contigs:
                SeqIO.write(kept_contigs, output_fasta, "fasta")

            qc_log.append({
                "assembly": assembly_id,
                "input_contigs": len(list(SeqIO.parse(input_fasta, "fasta"))),
                "kept_contigs": len(kept_contigs),
                "removed_contigs": removed_contigs
            })

print(f"‚úÖ Genome QC filtering complete")
print(f"üìÅ Cleaned genomes saved to: {OUTPUT_DIR}")
print(f"üß¨ Total genomes processed: {len(qc_log)}")
```

```python
from Bio import SeqIO
import os

# Parameters
K = 6
INPUT_DIR = "/content/genomes_qc"
OUTPUT_DIR = "/content/kmers"
os.makedirs(OUTPUT_DIR, exist_ok=True)

def extract_kmers(sequence, k):
    return [sequence[i:i+k] for i in range(len(sequence) - k + 1)]

genome_count = 0

for file in os.listdir(INPUT_DIR):
    if file.endswith(".fna"):
        genome_id = file.replace(".fna", "")
        input_fasta = os.path.join(INPUT_DIR, file)
        output_kmer_file = os.path.join(OUTPUT_DIR, f"{genome_id}.kmers.txt")

        kmers = []

        for record in SeqIO.parse(input_fasta, "fasta"):
            seq = str(record.seq).upper()
            if len(seq) >= K:
                kmers.extend(extract_kmers(seq, K))

        # Save ordered k-mers
        with open(output_kmer_file, "w") as f:
            f.write(" ".join(kmers))

        genome_count += 1

print(f"‚úÖ k-mer extraction completed")
print(f"üß¨ Genomes processed: {genome_count}")
print(f"üìÅ k-mer files saved to: {OUTPUT_DIR}")
```
