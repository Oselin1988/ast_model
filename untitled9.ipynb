```python
import pandas as pd
import re

df = pd.read_excel("/content/original 1000 samples.xlsx")
df.columns = df.columns.str.strip()

order_col = "S/N"
assembly_col = "assembly"
organism_col = "organism group"
source_col = "isolation source"
type_col = "isolation type of pathogen"
ast_col = "AST phenotype"

# Sort by original order
df = df.sort_values(order_col)

# ðŸ”‘ GLOBAL forward-fill of genome metadata
df[[assembly_col, organism_col, source_col, type_col]] = (
    df[[assembly_col, organism_col, source_col, type_col]].ffill()
)
```

```python
records = []

current_phenotype = None
remaining = 0

for _, row in df.iterrows():
    val = row[ast_col]

    if pd.isna(val):
        continue

    val = str(val).strip()

    # Discard non-informative rows
    if val.lower() in ["expand all"] or val.lower().startswith("other"):
        continue

    # Phenotype header
    m = re.match(r"(Resistant|Intermediate|Susceptible)\s*\((\d+)\)", val, re.I)
    if m:
        current_phenotype = m.group(1).capitalize()
        remaining = int(m.group(2))
        continue

    # Antibiotic rows
    if current_phenotype and remaining > 0:
        records.append({
            "assembly": row[assembly_col],
            "organism": row[organism_col],
            "isolation_source": row[source_col],
            "isolation_type": row[type_col],
            "antibiotic": val,
            "phenotype": current_phenotype
        })
        remaining -= 1

ast_long = pd.DataFrame(records)

print("ast_long shape:", ast_long.shape)
print("Unique assemblies:", ast_long["assembly"].nunique())
```

```python
import pandas as pd
import re

df = pd.read_excel("/content/original 1000 samples.xlsx")
df.columns = df.columns.str.strip()
df = df.sort_values("S/N").reset_index(drop=True)

# Rows where a new assembly starts
assembly_rows = df[df["assembly"].notna()].index.tolist()

print("Assemblies detected:", len(assembly_rows))  # should be 771
```

```python
records = []

for i, start_idx in enumerate(assembly_rows):
    end_idx = assembly_rows[i + 1] if i + 1 < len(assembly_rows) else len(df)

    block = df.iloc[start_idx:end_idx]

    assembly = block.iloc[0]["assembly"]
    organism = block.iloc[0]["organism group"]
    source = block.iloc[0]["isolation source"]
    iso_type = block.iloc[0]["isolation type of pathogen"]

    current_phenotype = None
    remaining = 0

    for _, row in block.iterrows():
        val = row["AST phenotype"]
        if pd.isna(val):
            continue

        val = str(val).strip()

        if val.lower() in ["expand all"] or val.lower().startswith("other"):
            continue

        m = re.match(r"(Resistant|Intermediate|Susceptible)\s*\((\d+)\)", val, re.I)
        if m:
            current_phenotype = m.group(1).capitalize()
            remaining = int(m.group(2))
            continue

        if current_phenotype and remaining > 0:
            records.append({
                "assembly": assembly,
                "organism": organism,
                "isolation_source": source,
                "isolation_type": iso_type,
                "antibiotic": val,
                "phenotype": current_phenotype
            })
            remaining -= 1

ast_long = pd.DataFrame(records)

print("ast_long shape:", ast_long.shape)
print("Unique assemblies:", ast_long["assembly"].nunique())
```

```python
import pandas as pd
import re

df = pd.read_excel("/content/original 1000 samples.xlsx")
df.columns = df.columns.str.strip()
df = df.sort_values("S/N").reset_index(drop=True)

# Identify genome-start rows (every Resistant(x) that is NOT an antibiotic)
genome_start_mask = df["AST phenotype"].astype(str).str.match(r"Resistant\s*\(\d+\)", na=False)

# Forward-fill assembly ONLY across genome starts
df.loc[genome_start_mask, "assembly"] = df.loc[genome_start_mask, "assembly"].ffill()

records = []

current_assembly = None
current_organism = None
current_source = None
current_type = None
current_phenotype = None
remaining = 0

for _, row in df.iterrows():

    # New genome begins here
    if genome_start_mask.loc[_]:
        current_assembly = row["assembly"]
        current_organism = row["organism group"]
        current_source = row["isolation source"]
        current_type = row["isolation type of pathogen"]
        current_phenotype = "Resistant"
        remaining = int(re.search(r"\((\d+)\)", row["AST phenotype"]).group(1))
        continue

    val = row["AST phenotype"]
    if pd.isna(val):
        continue

    val = str(val).strip()

    if val.lower() in ["expand all"] or val.lower().startswith("other"):
        continue

    m = re.match(r"(Intermediate|Susceptible)\s*\((\d+)\)", val, re.I)
    if m:
        current_phenotype = m.group(1).capitalize()
        remaining = int(m.group(2))
        continue

    if current_assembly and current_phenotype and remaining > 0:
        records.append({
            "assembly": current_assembly,
            "organism": current_organism,
            "isolation_source": current_source,
            "isolation_type": current_type,
            "antibiotic": val,
            "phenotype": current_phenotype
        })
        remaining -= 1

ast_long = pd.DataFrame(records)

print("ast_long shape:", ast_long.shape)
print("Unique assemblies:", ast_long["assembly"].nunique())
```
